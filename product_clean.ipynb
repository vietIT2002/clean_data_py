```python
import pandas as pd
import re

# Function to extract percentage values while keeping other details
def extract_percent_values(text):
    """
    Extracts percentage values while filtering out unwanted patterns.
    """
    if pd.isna(text):  # Handle missing values
        return []
    
    text = text.lower()

    # Extract percentage values
    all_percentages = re.findall(r'\d+\.?\d*%|\d+\.?\d*\s*(?:pct|percent)', text, re.IGNORECASE)

    return all_percentages

# Function to clean only unwanted patterns while keeping the rest intact
def clean_text(text):
    if pd.isna(text):  
        return ""

    text = text.lower()

    # Define only the unwanted patterns to remove
    exclude_patterns = [

        r'\bmới\s*\d+%?\b', r'\b\d+%\s*mới\b',  # Matches "mới 98%", "mới 120%", "mới 200%"
        r'\bnew\s*\d+%?\b', r'\b\d+%\s*new\b',  # Matches "new 98%", "new 125%"
        r'\bbrand\s*\d+%?\b', r'\b\d+%\s*brand\b',  # Matches "brand 98%", "98% brand"
        #r'\bsample\s*\d+%?\b', r'\b\d+%\s*sample\b',  # Matches "sample 98%", "98% sample"
        r'\bitem\s*\d+%?\b', r'\b\d+%\s*item\b',  # Matches "item 98%", "98% item"
        r'\bhàng\s*mới\s*\d+%?\b', r'\b\d+%\s*hàng\s*mới\b',  # Matches "hàng mới 98%", "98% hàng mới"
        r'\b\d+%\s*[\#&><?@!*]',  # Removes "98%" if followed by special characters
        r'\(\s*\+\-\s*\d+%\)', r'\+\-\s*\d+%',  # Matches "(+-5%)" or "+-5%"
        r'\(\s*\+/\-\s*\d+%\)', r'\+/\-\s*\d+%',  # Matches "(+/-5%)" or "+/-5%"
        #r'\.\s*\d+%\s*[#><?@&.*]'  # Matches ". 98% .", ". 98% $"
        r'(?<!\d)\.\s*\d+%\s*[#><?@&.*]'
    ]



    for pattern in exclude_patterns:
        text = re.sub(pattern, '', text, flags=re.IGNORECASE)

    # Remove extra spaces caused by removals
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply functions
df_dl["ProductList"] = df_dl["Product List"].apply(clean_text)
df_dl["Extracted_Percentages"] = df_dl["ProductList"].apply(extract_percent_values)

# Display the updated DataFrame
df_dl
```

```python
# Function to get position matching
def get_position_matching(row):
    """
    Tìm vị trí của các giá trị phần trăm đã trích xuất trong văn bản, đảm bảo khớp chính xác.
    Xử lý tất cả các lần xuất hiện, bao gồm cả trùng lặp, đồng thời đảm bảo các vị trí được nhận diện đúng.
    """
    extracted_percentages = row['Extracted_Percentages']  # List of extracted percentages
    if not extracted_percentages:
        return []

    text = str(row['ProductList']).lower()  # Convert text to lowercase for matching
    matches = []
    used_positions = set()
    
    # Ensure we match extracted percentages **in order**
    for percent in extracted_percentages:
        # Regex to find exact standalone percentage
        pattern = rf'(?<!\d){re.escape(percent)}(?!\d)'
        match_positions = [m.start() for m in re.finditer(pattern, text)]

        # Assign positions sequentially without duplicating previous matches
        for pos in match_positions:
            if pos not in used_positions:  # Avoid duplicate matches
                matches.append(pos)
                used_positions.add(pos)
                break  # Stop after finding the first valid match

    return sorted(matches)  # Return positions in order

# Apply function to DataFrame
df_dl['position_matching'] = df_dl.apply(get_position_matching, axis=1)


df_dl
```

```python
# Function to extract text using provided positions
def extract_text_from_row(row):
    text = row["ProductList"]
    positions = row["position_matching"]  # Predefined list of positions

    if not positions:
        return ""  # Return empty if no positions exist
    
    # Convert text into word tokens
    words = re.findall(r'\S+', text)
    word_positions = [text.index(word) for word in words]

    # Validate min and max positions within word_positions range
    min_pos = min(positions)
    max_pos = max(positions)

    # Find word indices corresponding to min_pos and max_pos safely
    min_word_idx = next((i for i, pos in enumerate(word_positions) if pos >= min_pos), 0)
    max_word_idx = next((i for i, pos in enumerate(word_positions) if pos >= max_pos), len(words) - 1)

    # Extract range from left window to right window
    left_window_idx = max(0, min_word_idx - 3)  # 2 words before min_pos
    right_window_idx = min(len(words), max_word_idx + 4)  # 2 words after max_pos

    # Extract and return the text within range
    return " ".join(words[left_window_idx:right_window_idx])

# Apply function to each row in the DataFrame
df_dl["extracted_text"] = df_dl.apply(extract_text_from_row, axis=1)


df_dl
```

```python
# Function to ensure '(' always has a matching ')' and keeps valid pairs correctly
def balance_parentheses(text):
    """
    Ensures that every '(' has a matching ')', removing only redundant ones.
    Valid pairs (containing text) are preserved, and misplaced words are corrected.
    """
    open_count = text.count('(')
    close_count = text.count(')')

    # Ensure balanced parentheses by removing excess unmatched ones
    while open_count > close_count:
        text = text[::-1].replace('(', '', 1)[::-1]  # Remove the last unmatched '('
        open_count -= 1

    while close_count > open_count:
        text = text.replace(')', '', 1)  # Remove the first unmatched ')'
        close_count -= 1

    # Ensure valid parentheses with content remain and stay in order
    text = re.sub(r'\(\s*([\w\d%]+.*?)\s*\)', r'(\1)', text)  # Keep valid content inside ()

    # Fix misplaced words outside parentheses (ensures spacing is maintained correctly)
    text = re.sub(r'(\S)\s+\(', r'\1 (', text)  # Ensure '(' follows a word properly
    text = re.sub(r'\)\s+(\S)', r') \1', text)  # Ensure ')' is followed by a space correctly

    # Remove redundant empty parentheses left at the end
    text = re.sub(r'\(\s*\)$', '', text)  # Removes lone '(' at the end

    return text.strip()

# Apply function to each row in the DataFrame
df_dl["balanced_text"] = df_dl["filtered_text"].apply(balance_parentheses)

df_dl=df_dl[['Product List','ProductList','filtered_text','balanced_text']]
df_dl
```

```python
def extract_percentages(text):
    matches = []  # Dùng list để giữ nguyên tất cả phần trăm xuất hiện
    text = re.sub(r'(\d+(?:\.\d+)?)\s+%', r'\1%', text)
    # Tìm tất cả các phần trăm có dạng '23%' hoặc '23.5%'
    direct_matches = re.findall(r'\d+\.?\d*\s?%', text)
    matches.extend(direct_matches)

    # Tìm các nhóm có dấu "/" hoặc "+"
    split_percentages = re.findall(r'([\d/+]+)(?:%|)', text)
    for group in split_percentages:
        if "/" in group or "+" in group:
            numbers = [int(num) for num in re.split(r'[/+]', group) if num.isdigit()]
            if sum(numbers) == 100:  # Chỉ thêm nếu tổng = 100%
                for num in numbers:
                    matches.append(f"{num}%")

    return matches

def find_percent_positions(text):
    positions = []

    # Tìm các vị trí phần trăm trong chuỗi
    for match in re.finditer(r'\d+\.\d+\s?%|\d+\s?%', text):
        positions.append(match.start())

    # Tìm các nhóm có dấu "/" và tính các vị trí cho các phần trăm tách ra
    for match in re.finditer(r'([\d/]+)(?:%|)', text):  
        if "/" in match.group(1):  
            numbers = [int(num) for num in match.group(1).split('/') if num.isdigit()]
            if sum(numbers) == 100:
                start_index = match.start(1)  
                current_pos = start_index
                for num in numbers:
                    if current_pos not in positions:  # Tránh trùng vị trí
                        positions.append(current_pos)
                    current_pos += len(str(num)) + 1  # Cập nhật vị trí

    # Sắp xếp theo vị trí trong chuỗi
    return sorted(positions)

# Áp dụng vào DataFrame của bạn
df_filtered_v['percent_values'] = df_filtered_v['ProductList'].apply(extract_percentages)
df_filtered_v['percent_positions'] = df_filtered_v['ProductList'].apply(find_percent_positions)

df_filtered_two = df_filtered_v

df_filtered_two
```

```python
# Function to ensure '(' always has a matching ')' and keeps valid pairs correctly
def balance_parentheses(text):
    """
    Ensures that every '(' has a matching ')', removing only redundant ones.
    Valid pairs (containing text) are preserved, and misplaced words are corrected.
    """
    open_count = text.count('(')
    close_count = text.count(')')

    # Ensure balanced parentheses by removing excess unmatched ones
    while open_count > close_count:
        text = text[::-1].replace('(', '', 1)[::-1]  # Remove the last unmatched '('
        open_count -= 1

    while close_count > open_count:
        text = text.replace(')', '', 1)  # Remove the first unmatched ')'
        close_count -= 1

    # Ensure valid parentheses with content remain and stay in order
    text = re.sub(r'\(\s*([\w\d%]+.*?)\s*\)', r'(\1)', text)  # Keep valid content inside ()

    # Fix misplaced words outside parentheses (ensures spacing is maintained correctly)
    text = re.sub(r'(\S)\s+\(', r'\1 (', text)  # Ensure '(' follows a word properly
    text = re.sub(r'\)\s+(\S)', r') \1', text)  # Ensure ')' is followed by a space correctly

    # Remove redundant empty parentheses left at the end
    text = re.sub(r'\(\s*\)$', '', text)  # Removes lone '(' at the end

    return text.strip()

# Apply function to each row in the DataFrame
# df_final_v["balanced_text"] = df_final_v["cleaned_text"].apply(balance_parentheses)

# df_final_v=df_final_v[['ProductList','cleaned_text','balanced_text', 'percent_positions']]
# df_final_v
# Function to remove empty parentheses from text
def remove_empty_parentheses(text):
    """
    Removes empty parentheses '()' from the text while preserving valid ones.
    """
    return re.sub(r'\(\s*\)', '', text).strip()

# # Apply function to each row in the DataFrame
# df_final_v["cleaned_text"] = df_final_v["balanced_text"].apply(remove_empty_parentheses)
# df_final_v=df_final_v[['ProductList','cleaned_text', 'percent_positions']]
# # df_dl.to_excel('df_DL.xlsx', engine='openpyxl')
# if df_final_v.empty:
#     print("DataFrame df_final hiện đang rỗng")
# else:
#     df_final_v
try:
    # Nếu df_final_v tồn tại thì xử lý
    df_final_v["balanced_text"] = df_final_v["cleaned_text"].apply(balance_parentheses)
    df_final_v = df_final_v[['ProductList', 'cleaned_text', 'balanced_text', 'percent_positions']]
    
    df_final_v["cleaned_text"] = df_final_v["balanced_text"].apply(remove_empty_parentheses)
    df_final_v = df_final_v[['ProductList', 'cleaned_text', 'percent_positions']]

    if df_final_v.empty:
        print("DataFrame df_final_v hiện đang rỗng.")
    else:
        df_final_v

except NameError:
    print("DataFrame df_final_v chưa được tạo ra")
    df_final_v=df_filtered
df_final_v
```

```python
# Function to ensure '(' always has a matching ')' and keeps valid pairs correctly
def balance_parentheses(text):
    """
    Ensures that every '(' has a matching ')', removing only redundant ones.
    Valid pairs (containing text) are preserved, and misplaced words are corrected.
    """
    open_count = text.count('(')
    close_count = text.count(')')

    # Ensure balanced parentheses by removing excess unmatched ones
    while open_count > close_count:
        text = text[::-1].replace('(', '', 1)[::-1]  # Remove the last unmatched '('
        open_count -= 1

    while close_count > open_count:
        text = text.replace(')', '', 1)  # Remove the first unmatched ')'
        close_count -= 1

    # Ensure valid parentheses with content remain and stay in order
    text = re.sub(r'\(\s*([\w\d%]+.*?)\s*\)', r'(\1)', text)  # Keep valid content inside ()

    # Fix misplaced words outside parentheses (ensures spacing is maintained correctly)
    text = re.sub(r'(\S)\s+\(', r'\1 (', text)  # Ensure '(' follows a word properly
    text = re.sub(r'\)\s+(\S)', r') \1', text)  # Ensure ')' is followed by a space correctly

    # Remove redundant empty parentheses left at the end
    text = re.sub(r'\(\s*\)$', '', text)  # Removes lone '(' at the end

    return text.strip()

# Apply function to each row in the DataFrame
df_filtered_100["balanced_text"] = df_filtered_100["cleaned_text"].apply(balance_parentheses)

df_filtered_100=df_filtered_100[['ProductList','cleaned_text','balanced_text', 'percent_positions']]
df_filtered_100
def remove_empty_parentheses(text):
    """
    Removes empty parentheses '()' from the text while preserving valid ones.
    """
    return re.sub(r'\(\s*\)', '', text).strip()

# Apply function to each row in the DataFrame
df_filtered_100["cleaned_text"] = df_filtered_100["balanced_text"].apply(remove_empty_parentheses)
df_filtered_100=df_filtered_100[['ProductList','cleaned_text', 'percent_positions']]
# df_dl.to_excel('df_DL.xlsx', engine='openpyxl')
df_filtered_100
```

### Tìm vị trí xuất hiện từ khóa trong chuỗi, nếu không có thì lấy vị trí của số nguyên đầu tiên

```python
def find_first_integer_position(text):
    """
    Tìm vị trí xuất hiện đầu tiên của số nguyên trong chuỗi.
    Trả về độ dài chuỗi nếu không có số nào.
    """
    if not isinstance(text, str):
        return -1
    
    match = re.search(r'\d+', text)
    return match.start() if match else len(text)

def max_keyword_or_integer_position(text, keywords):
    """
    Tìm vị trí xuất hiện cao nhất trong các từ khóa. Nếu không có từ khóa,
    trả về vị trí đầu tiên của số nguyên trong chuỗi. Nếu không có số nguyên,
    trả về độ dài chuỗi.
    """
    if not isinstance(text, str):
        return -1
        
    # Tìm vị trí xuất hiện đầu tiên của từng keyword
    positions = [text.find(k) for k in keywords]
    valid_positions = [pos for pos in positions if pos != -1]

    if valid_positions:
        return max(valid_positions)
    else:
        return find_first_integer_position(text)


keywords = ['khổ:', "khổ",'k=','k:', 'size', 'k.', 'w:','w.', 'kt:','kt.']

# Áp dụng hàm cho từng dòng
df['MaxKeywordPos'] = df['ProductList'].apply(lambda x: max_keyword_or_integer_position(x, keywords))
df
```

```python
#lọc những dòng có MaxKeywordPos = 1 và 0 cắt những cụm từ số nguyên ở đầu chuỗi nối vào sau giá trị của cột TapNham
def append_leading_number_to_tapnham(row):
    pos = row['MaxKeywordPos']
    product_text = row['ProductList']
    tapnham_text = row['TapNham']
    
    if pos in [0, 1]:
        # Tìm số nguyên ở đầu (không quan tâm ký tự sau)
        match = re.match(r'^(\d+)', product_text.strip())
        if match:
            number = match.group(1)
            # Cắt phần số khỏi đầu chuỗi bằng cách dùng độ dài
            product_new = product_text[len(number):].lstrip()
            
            # Nối vào TapNham
            if pd.isna(tapnham_text):
                tapnham_new = number
            else:
                tapnham_new = f"{tapnham_text.strip()} {number}"
            
            return pd.Series([tapnham_new, product_new])
    
    # Trường hợp không thỏa
    return pd.Series([tapnham_text, product_text])


# Áp dụng hàm vào DataFrame
df[['TapNham', 'ProductList']] = df.apply(append_leading_number_to_tapnham, axis=1)
# Áp dụng hàm cho từng dòng
# df['MaxKeywordPos'] = df['ProductList'].apply(lambda x: max_keyword_or_integer_position(x, keywords))
```

```python
def xu_ly_max_keyword_pos_1(df):
    """
    Xử lý các dòng có MaxKeywordPos = 1:
    - Nếu giá trị đơn trong cột Product trùng với ký tự cuối của TapNham,
      chỉ cần loại bỏ khỏi Product
    - Ngược lại, di chuyển ký tự đó vào cuối TapNham và loại bỏ khỏi Product
    - Đồng thời loại bỏ giá trị tương ứng ở đầu chuỗi của cột ProductList
    
    Tham số:
        df: DataFrame có các cột 'Product', 'TapNham', 'ProductList' và 'MaxKeywordPos'
        
    Trả về:
        DataFrame đã được xử lý
    """
    # Tạo bản sao để tránh thay đổi DataFrame gốc
    df_ket_qua = df.copy()
    
    # Lọc các dòng có MaxKeywordPos = 1
    mask = df_ket_qua['MaxKeywordPos'] == 1
    
    for idx in df_ket_qua[mask].index:
        gia_tri_product = df_ket_qua.at[idx, 'Product']
        gia_tri_tapnham = df_ket_qua.at[idx, 'TapNham']
        gia_tri_productlist = df_ket_qua.at[idx, 'ProductList']
        
        # Kiểm tra xem Product có phải là một ký tự đơn không
        if isinstance(gia_tri_product, str) and len(gia_tri_product.strip()) == 1:
            ky_tu_product = gia_tri_product.strip()
            
            # Xử lý cột TapNham
            if isinstance(gia_tri_tapnham, str) and gia_tri_tapnham:
                # Kiểm tra xem ký tự cuối của TapNham có trùng với ký tự Product không
                if gia_tri_tapnham[-1] == ky_tu_product:
                    # Chỉ loại bỏ khỏi Product mà không sửa đổi TapNham
                    df_ket_qua.at[idx, 'Product'] = ''
                else:
                    # Di chuyển ký tự vào cuối TapNham
                    df_ket_qua.at[idx, 'TapNham'] = gia_tri_tapnham + ky_tu_product
                    df_ket_qua.at[idx, 'Product'] = ''
            else:
                # Nếu TapNham trống hoặc không phải chuỗi, chỉ cần di chuyển ký tự
                df_ket_qua.at[idx, 'TapNham'] = str(gia_tri_tapnham or '') + ky_tu_product
                df_ket_qua.at[idx, 'Product'] = ''
            
            # Xử lý cột ProductList - loại bỏ ký tự đầu tiên nếu trùng với ký tự Product
            if isinstance(gia_tri_productlist, str) and gia_tri_productlist:
                if gia_tri_productlist.startswith(ky_tu_product):
                    # Loại bỏ ký tự đầu tiên
                    df_ket_qua.at[idx, 'ProductList'] = gia_tri_productlist[1:]
                elif gia_tri_productlist.startswith(' ' + ky_tu_product):
                    # Trường hợp bắt đầu bằng khoảng trắng + ký tự
                    df_ket_qua.at[idx, 'ProductList'] = gia_tri_productlist[2:]
    
    return df_ket_qua

# Ví dụ sử dụng
df = xu_ly_max_keyword_pos_1(df)
df
```

```python
def xu_ly_max_keyword_pos_2(df):
    """
    Xử lý các dòng có MaxKeywordPos = 1, 2 hoặc 3:
    - Kiểm tra 3 ký tự đầu tiên của cột Specifications
    - Nếu có ít nhất 1 trong 3 ký tự đầu tiên là số, thì:
      + Cắt chuỗi từ đầu đến khoảng trắng hoặc dấu đặc biệt (-, ", #, &)
      + Di chuyển chuỗi đó vào cuối TapNham
      + Loại bỏ chuỗi đó khỏi đầu Specifications và ProductList
    
    Tham số:
        df: DataFrame có các cột 'Specifications', 'TapNham', 'ProductList' và 'MaxKeywordPos'
        
    Trả về:
        DataFrame đã được xử lý
    """
    # Tạo bản sao để tránh thay đổi DataFrame gốc
    df_ket_qua = df.copy()
    
    # Lọc các dòng có MaxKeywordPos = 1, 2 hoặc 3
    mask = df_ket_qua['MaxKeywordPos'].isin([1, 2, 3])
    
    for idx in df_ket_qua[mask].index:
        specs = df_ket_qua.at[idx, 'Specifications']
        tapnham = df_ket_qua.at[idx, 'TapNham']
        product_list = df_ket_qua.at[idx, 'ProductList']
        
        # Kiểm tra xem Specifications có phải là chuỗi không
        if not isinstance(specs, str) or not specs:
            continue
            
        # Kiểm tra 3 ký tự đầu tiên có chứa số không
        first_three = specs[:min(3, len(specs))]
        if any(char.isdigit() for char in first_three):
            # Tìm vị trí của khoảng trắng hoặc dấu đặc biệt đầu tiên
            match = re.search(r'[ \&]', specs)
            
            if match:
                cut_position = match.start()
                # Chuỗi cần cắt và di chuyển
                cut_string = specs[:cut_position]
                
                # Cập nhật TapNham
                if isinstance(tapnham, str):
                    df_ket_qua.at[idx, 'TapNham'] = tapnham + cut_string
                else:
                    df_ket_qua.at[idx, 'TapNham'] = str(tapnham or '') + cut_string
                
                # Cập nhật Specifications
                df_ket_qua.at[idx, 'Specifications'] = specs[cut_position:].lstrip()
                
                # Cập nhật ProductList nếu có
                if isinstance(product_list, str) and product_list:
                    # Kiểm tra xem ProductList có bắt đầu bằng chuỗi cần cắt không
                    if product_list.startswith(cut_string):
                        df_ket_qua.at[idx, 'ProductList'] = product_list[len(cut_string):].lstrip()
                    # Trường hợp có khoảng trắng ở đầu
                    elif product_list.lstrip().startswith(cut_string):
                        leading_spaces = len(product_list) - len(product_list.lstrip())
                        df_ket_qua.at[idx, 'ProductList'] = product_list[:leading_spaces] + product_list[leading_spaces + len(cut_string):].lstrip()
            else:
                # Nếu không tìm thấy khoảng trắng hoặc dấu đặc biệt, di chuyển toàn bộ chuỗi
                if isinstance(tapnham, str):
                    df_ket_qua.at[idx, 'TapNham'] = tapnham + specs
                else:
                    df_ket_qua.at[idx, 'TapNham'] = str(tapnham or '') + specs
                
                df_ket_qua.at[idx, 'Specifications'] = ''
                
                # Cập nhật ProductList nếu có
                if isinstance(product_list, str) and product_list:
                    if product_list.startswith(specs):
                        df_ket_qua.at[idx, 'ProductList'] = product_list[len(specs):].lstrip()
                    elif product_list.lstrip().startswith(specs):
                        leading_spaces = len(product_list) - len(product_list.lstrip())
                        df_ket_qua.at[idx, 'ProductList'] = product_list[:leading_spaces] + product_list[leading_spaces + len(specs):].lstrip()
    
    return df_ket_qua

# Ví dụ sử dụng
df = xu_ly_max_keyword_pos_2(df)
df
```
